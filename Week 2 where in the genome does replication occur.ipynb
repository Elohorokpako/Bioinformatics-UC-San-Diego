### Week 2: Where in the genome does replication occur?

#### Exercise Break: 

Give all values of Skew [i] (GAGCCACCGCGATA) for i ranging from 0 to 14.

Sample Input:

     CATGGGCATCGGCCATACGCC

Sample Output:

     0 -1 -1 -1 0 1 2 1 1 1 0 1 2 1 0 0 0 0 -1 0 -1 -2

```python
#this function tracks the changes in the genome and assigns a value of 0 at the beginning. looping through the genome,
#if the base is C, -1 is deducted, if it is G, +1 is computed, if it is A, or T, it stays as it is.
def SkewArray(Genome):
    skew = [0]
    score = {"A":0, "T":0, "C":-1, "G":1}
    for i in range(1,len(Genome)+1):
            skew.append(score[Genome[i-1]] + skew[i-1])
    return skew
```

### Question 1

Minimum Skew Problem: Find a position in a genome where the skew diagram attains a minimum.

Input: A DNA string Genome.

Output: All integer(s) i minimizing Skewi (Genome) among all values of i (from 0 to |Genome|).

Code Challenge: Solve the Minimum Skew Problem.

Sample Input:

TAAAGACTGCCGAGAGGCCAACACGAGTGCTAGAACGAGGGGCGTAAACGCGGGTCCGAT

Sample Output:

11 24

```python
#this is a continuation of the first fucntion. In addition, this function returns the index of the base with the
#lowest score.
def skew_array(genome):
    score = {"A": 0, "T": 0, "C": -1, "G": 1}
    skew = [0] * (len(genome) + 1)

    for i in range(1, len(genome) + 1):
        skew[i] = score[genome[i - 1]] + skew[i - 1]

    min_skew = min(skew)
    answer = [index for index, value in enumerate(skew) if value == min_skew]
    return answer
```

```python
print(*skew_array("TAAAGACTGCCGAGAGGCCAACACGAGTGCTAGAACGAGGGGCGTAAACGCGGGTCCGAT"))
```

### Question 2

Hamming Distance Problem: Compute the Hamming distance between two strings.

Input: Two strings of equal length.

Output: The Hamming distance between these strings.

Sample Input:

GGGCCGTTGGT

GGACCGTTGAC

Sample Output:

3

```python
#this function computes the difference between two genomes (a, and b) at each position and returns the sum of the
#differences between them
def hamming_distance(a,b):
    answer=sum([1 for i in range(len(a)) if a[i]!=b[i]])
    return answer
```

### Question 3
Approximate Pattern Matching Problem: Find all approximate occurrences of a pattern in a string.

Input: Strings Pattern and Text along with an integer d.

Output: All starting positions where Pattern appears as a substring of Text with at most d mismatches.

Sample Input:

ATTCTGGA

CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAAT

3

Sample Output:

6 7 26 27

```python
#this function is a continuation of the hamming_distance function. It computes the hamming distance of each kmer
#and the pattern provided and returns the index of the kmers whose hamming distance is <= d
def hamming_distance(a,b):
    answer=sum([1 for i in range(len(a)) if a[i]!=b[i]])
    return answer
def pattern_matching(pattern, genome,d): 
    g=len(genome)
    p=len(pattern)
    kmer=[0]*(g-p+1)
    for i in range(g-p+1):
        kmer[i]=genome[i:i+p]
    positions=[]
    positions+=[index for index,val in enumerate(kmer) if hamming_distance(val,pattern)<=d]
    return positions
```

```python
print(*pattern_matching("ATTCTGGA","CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAAT",3))
```

### Question 5

Code Challenge: Solve the Frequent Words with Mismatches Problem.

Input: A string Text as well as integers k and d. (You may assume k ≤ 12 and d ≤ 3.)

Output: All most frequent k-mers with up to d mismatches in Text.

```python
def hamming_distance(a,b):
    answer=sum([1 for i in range(len(a)) if a[i]!=b[i]])
    return answer

def neighbors(pattern, d):
    if d == 0:
        return {pattern}
    if len(pattern) == 1:
        return {'A', 'C', 'G', 'T'}
    neighborhood = set()
    suffix_neighbors = neighbors(pattern[1:], d)
    for text in suffix_neighbors:
        if hamming_distance(pattern[1:], text) < d:
            for nucleotide in ['A', 'C', 'G', 'T']:
                neighborhood.add(nucleotide + text)
        else:
            neighborhood.add(pattern[0] + text)
    return neighborhood

def frequent_words_with_mismatches(text, k, d):
    patterns = []
    freq_map = {}
    n = len(text)
    
    for i in range(n - k + 1):
        pattern = text[i:i+k]
        neighborhood = neighbors(pattern, d)
        
        for neighbor in neighborhood:
            if neighbor not in freq_map:
                freq_map[neighbor] = 1
            else:
                freq_map[neighbor] += 1
    
    m = max(freq_map.values())

    for pattern in freq_map:
        if freq_map[pattern] == m:
            patterns.append(pattern)
    
    return patterns
```

```python
#the function frequent_words_with_mismatches_and_rc is a modification of the frequent word problem. This code instead
#returns the sum of the frequent word and it's complement that have the highest frequency in combination.
def reverse_complement (dna):
    complement=list(dna)
    for i in range(len(dna)):
        if complement[i]=="A":
            complement[i]="T"
        elif complement[i]=="T":
            complement[i]="A"
        elif complement[i]=="C":
            complement[i]="G"
        elif complement[i]=="G":
            complement[i]="C"
    return "".join(complement[::-1])

def hamming_distance(a,b):
    answer=sum([1 for i in range(len(a)) if a[i]!=b[i]])
    return answer

def neighbors(pattern, d):
    if d == 0:
        return {pattern}
    if len(pattern) == 1:
        return {'A', 'C', 'G', 'T'}
    neighborhood = set()
    suffix_neighbors = neighbors(pattern[1:], d)
    for text in suffix_neighbors:
        if hamming_distance(pattern[1:], text) < d:
            for nucleotide in ['A', 'C', 'G', 'T']:
                neighborhood.add(nucleotide + text)
        else:
            neighborhood.add(pattern[0] + text)
    return neighborhood

def frequent_words_with_mismatches_and_rc(text, k, d):
    patterns = []
    freq_map = {}
    n = len(text)
    
    for i in range(n - k + 1):
        pattern = text[i:i+k]
        neighborhood = neighbors(pattern, d)
        
        for neighbor in neighborhood:
            if neighbor not in freq_map:
                freq_map[neighbor] = 1
            else:
                freq_map[neighbor] += 1
    
    #m = max(freq_map.values())

    answers=[]       
    n=[]
    max_n=''
    for key in freq_map:
        if reverse_complement(key) not in freq_map:
            pass
        else:
            n.append(freq_map[key] +freq_map[reverse_complement(key)])
    max_n=max(n)
    for keys in freq_map:
        if reverse_complement (keys) in freq_map:
            if freq_map[keys] +freq_map[reverse_complement(keys)]==max_n:
                answers.append(keys)
                
    return answers
```

```python
def skew_max(genome):
    score = {"A": 0, "T": 0, "C": -1, "G": 1}
    skew = [0] * (len(genome) + 1)

    for i in range(1, len(genome) + 1):
        skew[i] = score[genome[i - 1]] + skew[i - 1]

    max_skew = max(skew)
    answer = [index for index, value in enumerate(skew) if value == max_skew]
    return answer
```
